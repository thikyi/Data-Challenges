# Part 1 ‐ Exploratory data analysis

The attached logins.json file contains (simulated) timestamps of user logins in a particular
geographic location.</br> Aggregate these login counts based on 15minute time intervals, and
visualize and describe the resulting time series of login counts in ways that best characterize the
underlying patterns of the demand. Please report/illustrate important features of the demand,
such as daily cycles. If there are data quality issues, please report them.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_logins = pd.read_json('logins.json')
df_logins
```

Aggregate these login counts based on 15-minute time intervals 

```python
df_logins.index = pd.to_datetime(df_logins['login_time'])
df_logins_15mins = df_logins.resample('15T').count()
```

```python
df_logins_15mins
```

```python
df_logins_15mins.rename(columns={'login_time': 'login_count'}, inplace=True)
df_logins_15mins
```

```python
df_logins_15mins = df_logins_15mins.reset_index()
```

```python
def categorize_time(timestamp):
    hour = timestamp.hour  
    if 0 <= hour < 12:
        return "AM"
    else:
        return "PM"
```

```python
df_logins_15mins["time_type"] = df_logins_15mins["login_time"].apply(categorize_time)
df_logins_15mins
```

```python
df_logins_15mins['mth'] = df_logins_15mins['login_time'].dt.month
df_logins_15mins['day'] = df_logins_15mins['login_time'].dt.day
df_logins_15mins['hr'] = df_logins_15mins['login_time'].dt.hour
df_logins_15mins['yr'] = df_logins_15mins['login_time'].dt.year
df_logins_15mins['wdno']=df_logins_15mins['login_time'].dt.dayofweek
df_logins_15mins.loc[:, 'wd'] = df_logins_15mins['login_time'].dt.strftime('%A')
```

```python
df_logins_15mins = df_logins_15mins.set_index(df_logins_15mins.columns[0])
df_logins_15mins
```

##### Time series of login counts

```python

plt.figure(figsize=(20, 6))
plt.plot(df_logins_15mins.index, df_logins_15mins.login_count, label='Login Counts')
plt.title('Time Series of Login Counts (15-minute intervals)')
plt.xlabel('Time')
plt.ylabel('Login Counts')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
df_am = df_logins_15mins[df_logins_15mins['time_type'] == 'AM']
df_pm = df_logins_15mins[df_logins_15mins['time_type'] == 'PM']


plt.figure(figsize=(20, 6))

# Since we have no AM data in the snippet, it will not be plotted, but the code is here for completeness.
if not df_am.empty:
    plt.plot(df_am.index, df_am['login_count'], color='blue', label='AM')

plt.plot(df_pm.index, df_pm['login_count'], color='red', label='PM')

plt.title('Login Counts Time Series - AM vs. PM')
plt.xlabel('Time')
plt.ylabel('Login Count')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout() 

# Show the plot
plt.show()
```

```python
hourly_average = df_logins_15mins.groupby(df_logins_15mins['hr'])['login_count'].mean()

# Plot the average login counts by hour
plt.figure(figsize=(14, 6))
hourly_average.plot(kind='bar', title='Average Login Counts by Hour')
plt.xlabel('Hour of the Day')
plt.ylabel('Average Login Counts')
plt.xticks(rotation=0)
plt.grid(True)
plt.show()
```

```python
dayofweek_average = df_logins_15mins.groupby(df_logins_15mins['wd'])['login_count'].mean()

# Plot the average login counts by hour
plt.figure(figsize=(14, 6))
dayofweek_average.plot(kind='bar', title='Average Login Counts by day of the week')
plt.xlabel('day of the Week')
plt.ylabel('Average Login Counts')
plt.xticks(rotation=0)
plt.grid(True)
plt.show()
```

##### Since the dataset might cover a long period, let's also plot a shorter, more detailed period to observe daily cycles more clearly

```python
plt.figure(figsize=(20, 6))
# Selecting a one-week period for detailed visualization
one_week = df_logins_15mins.loc['1970-01-02':'1970-01-08']
plt.plot(one_week.index, one_week.login_count, label='Login Counts', color='orange')
plt.title('Detailed View: One Week of Login Counts (15-minute intervals)')
plt.xlabel('Time')
plt.ylabel('Login Counts')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

#### Finding

<b>Peak Times</b>: Consistent peaks during the detailed week suggest times of high user activity, likely in the morning and evening of the day. Saturday has the highest login traffic.</br>
<b>Low Activity Periods</b>: There are clear periods of low activity, typically early morning 5AM to 9 AM.</br>
<b>Data Quality</b>: No major data quality issues like missing data are immediately apparent, but minor issues (e.g., duplicate entries, incorrect timestamps) may require further analysis to detect.</br>

```python
# Return the performance metrics for the models
rf_performance = (rf_accuracy, rf_report)
logreg_performance = (logreg_accuracy, logreg_report)
svm_performance = (svm_accuracy, svm_report)

rf_performance, logreg_performance, svm_performance
```

```python
def extract_metrics(performance):
    accuracy, metrics_str = performance
    metrics_split = metrics_str.split(';')
    class_0_metrics = [float(metric) for metric in metrics_split[0].split(',')]
    class_1_metrics = [float(metric) for metric in metrics_split[1].split(',')]
    return accuracy, class_0_metrics, class_1_metrics

```

```python
import re

# Corrected function to extract metrics from the detailed report string
def extract_metrics_from_report(report):
    # Find all numeric values in the report
    numeric_values = re.findall(r"[\d.]+", report)
    # Convert found strings to floats
    metrics = list(map(float, numeric_values))
    # Assuming the order of metrics is precision, recall, f1-score, and support for each class
    # and then accuracy, macro avg, and weighted avg at the end
    # Extracting the necessary values: 
    accuracy = metrics[-7]  # The accuracy is typically at the -7th position from the end
    precision_class_0 = metrics[0]
    recall_class_0 = metrics[1]
    f1_score_class_0 = metrics[2]
    support_class_0 = metrics[3]
    precision_class_1 = metrics[4]
    recall_class_1 = metrics[5]
    f1_score_class_1 = metrics[6]
    support_class_1 = metrics[7]
    return accuracy, [precision_class_0, recall_class_0, f1_score_class_0, support_class_0], [precision_class_1, recall_class_1, f1_score_class_1, support_class_1]
```

```python
rf_accuracy, rf_class_0_metrics, rf_class_1_metrics = extract_metrics_from_report(rf_performance[1])
logreg_accuracy, logreg_class_0_metrics, logreg_class_1_metrics = extract_metrics_from_report(logreg_performance[1])
svm_accuracy, svm_class_0_metrics, svm_class_1_metrics = extract_metrics_from_report(svm_performance[1])

```
